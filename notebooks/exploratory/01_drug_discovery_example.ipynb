This notebook demonstrates basic drug discovery workflows including:

## 1. Molecular Property Prediction

Predict physicochemical and drug-likeness properties.

```python
# Example SMILES strings
example_smiles = [
    'CC(C)Cc1ccc(cc1)C(C)C(O)=O',  # Ibuprofen
    'CC(=O)Oc1ccccc1C(=O)O',        # Aspirin
]

predictor = MolecularPropertyPredictor()
properties = predictor.predict_properties(example_smiles)

print("Predicted Properties:")
print(properties)
```
